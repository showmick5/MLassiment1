```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import sklearn
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("CarPrice_Assignment.csv")
df.head()
```

```python
df.columns
```

```python
df['car_ID'].unique()
```

```python
df.drop("car_ID",axis=1,inplace=True)
```

```python
df.head()
```

```python
df['symboling'].unique()
```

```python
len(df['CarName'].unique())
```

```python
df.drop('CarName',axis=1, inplace=True)
```

```python
df['fueltype'].unique()
```

```python
#one hot encoding
fuel = pd.get_dummies(df['fueltype'], prefix = 'fueltype' ,drop_first=True)
type(fuel)
```

```python
fuel
```

```python
df = pd.concat((df,fuel),axis=1)
```

```python

df.drop('fueltype',axis=1,inplace=True)
df.head()
```

```python
df['aspiration'].unique()
```

```python
#One hot encoding of aspiration
aspi = pd.get_dummies(df['aspiration'], prefix = 'aspiration' ,drop_first=True)
df = pd.concat((df,aspi),axis=1)
df.drop('aspiration',axis=1,inplace=True)
df.head()
```

```python
df['doornumber'].unique()
```

```python
#one hot encdoing of doornumber
door = pd.get_dummies(df['doornumber'], prefix = 'doornumber' ,drop_first=True)
df = pd.concat((df,door),axis=1)
df.drop('doornumber',axis=1,inplace=True)
df.head()
```

```python
df['carbody'].unique()
```

```python
la = LabelEncoder()
```

```python
df['carbody'] = la.fit_transform(df['carbody'])
```

```python
df.head()
```

```python
df['drivewheel'].unique()
```

```python
#one hot encdoing of drivewheel
drivewheel = pd.get_dummies(df['drivewheel'], prefix = 'drivewheel' ,drop_first=True)
df = pd.concat((df,drivewheel),axis=1)
df.drop('drivewheel',axis=1,inplace=True)
df.head()
```

```python
df['enginelocation'].unique()
```

```python
#one hot encdoing of drivewheel
el= pd.get_dummies(df['enginelocation'], prefix = 'enginelocation', drop_first=True)    
df = pd.concat((df,el),axis=1)
df.drop('enginelocation',axis=1,inplace=True)
df.head()
```

```python
df['enginetype'].unique()
```

```python
df['enginetype'] = la.fit_transform(df['enginetype'])
df.head()
```

```python
df['cylindernumber'].unique()
```

```python
df['cylindernumber'] = la.fit_transform(df['cylindernumber'])
df.head()
```

```python
df.columns
```

```python
df.drop('enginesize',axis=1 ,inplace=True)
df.head()
```

```python
df['fuelsystem'].unique()
```

```python
df['fuelsystem'] = la.fit_transform(df['fuelsystem'])
df.head()
```

```python
df.drop(['boreratio','stroke','compressionratio','peakrpm', 'citympg', 'highwaympg'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['horsepower'].unique()
```

```python
df.drop('horsepower',axis=True,inplace=True)
```

```python
set(df['curbweight'].unique())
```

```python
df['curbweight'][df['curbweight']>3000]=0
df['curbweight'][df['curbweight']>2500]=1
df['curbweight'][df['curbweight']>2000]=2
df['curbweight'][df['curbweight']>1500]=3
```

```python
df.head()
```

```python
df.drop(['wheelbase','carlength','carwidth','carheight'],axis=True , inplace=True)
```

```python
df.head()
```

```python
X = df.drop("price",axis=1)
```

```python
X
```

```python
y= df['price']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
sc = StandardScaler()
```

```python
X_scale = sc.fit_transform(X)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X_scale, y, test_size=0.2 , random_state=322)
```

```python
liner_model = LinearRegression()
```

```python
liner_model.fit(x_train,y_train)
```

```python
liner_model.score(x_test, y_test)
```

```python
liner_model.score(x_train,y_train)
```
