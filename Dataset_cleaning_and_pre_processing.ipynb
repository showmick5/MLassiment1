```python
import pandas as pd
```

```python
df=pd.read_csv("googleplaystore.csv")
df.head()
```

```python
print(df.shape)
print(df.dtypes)
```

```python
#checking null value
df.isnull().sum()
```

```python
#Drop null value of multiple columns
df.dropna(subset=['Content Rating','Type','Current Ver','Android Ver'],inplace=True)
```

```python
df.isnull().sum()
```

```python
#checking duplicated value
df[df.duplicated() == True].count()
```

```python
#Droping duplicates
df=df.drop_duplicates()
```

```python
df[df.duplicated() == True].count()
```

```python
df.shape
```

```python
#Droping App as it is not important
df.drop('App', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Category'].unique()
```

```python
df['Category'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder 
label_encoder = LabelEncoder()
```

```python
#Label encoder of Category columns
df['Category_label'] = label_encoder.fit_transform(df['Category'])
df['Category_label'].head()
```

```python
df.head()
```

```python
df['Rating'].isna().sum()
```

```python
#Replacing null value with mean value
df['Rating'].fillna(df['Rating'].mean(), inplace=True)
df['Rating'].isna().sum()
```

```python
#Finding last different value of size
find_mem = df['Size'].apply(lambda x: x[-1])
set(find_mem)
```

```python
#replaceing Find_mem set 
df['Size'] = df['Size'].apply(lambda x: x.replace("M", "") if x[-1] == 'M' else x)
df['Size'] = df['Size'].apply(lambda x: x.replace("k", "00") if x[-1] == 'k' else x)
df['Size'] = df['Size'].apply(lambda x: x.replace("Varies with device", "0") if x == 'Varies with device' else x)
```

```python
#Converting size into float
df['Size'] = df['Size'].apply(lambda x: float(x))
```

```python
df.head()
```

```python
#converting reviews object to int value
df['Reviews'] = df['Reviews'].apply(lambda x: int(x))
```

```python
df.dtypes
```

```python
#checking if any value other that (+) existed
find_mem = df['Installs'].apply(lambda x: x[-1])
set(find_mem)
```

```python
#replace (+)
df['Installs'] = df['Installs'].apply(lambda x: x.replace("+", "") if x[-1] == '+' else x)
```

```python
df.head()
```

```python
#replacing (,)  and converting install into int
df['Installs']=df['Installs'].str.replace(",", "") 
df['Installs'] = df['Installs'].apply(lambda x: int(x))
df
```

```python
df['Type'].unique()
```

```python
#one hot encoding of Type featue
df['Type']=pd.get_dummies(df['Type'],drop_first=True)
```

```python
df.head()
```

```python
df['Type'].unique()
```

```python
df['Price'].unique()
```

```python
#Replacing $ 
df['Price']=df['Price'].str.replace("$", "") 
```

```python
#converting Price into float
df['Price'] = df['Price'].apply(lambda x: float(x))
```

```python
df.head()
```

```python
df['Content Rating'].unique()
```

```python
#Droping unrated rows
df['Content Rating']=df['Content Rating'][df['Content Rating'] != 'Unrated']
```

```python
#label encoding of Content Rating
df['Content Rating'] = label_encoder.fit_transform(df['Content Rating'])
df.head()
```

```python
df['Genres'].unique()
```

```python
len(df['Genres'].unique())
```

```python
#Lable encoding of Genres_label
df['Genres_label'] = label_encoder.fit_transform(df['Genres'])
df.head()
```

```python
#Spliting Last Updated into multiple columns
Time=df['Last Updated'].str.split(expand=True)
```

```python
#assinging month , day and year to respective columns
df['Month']=Time[0]
df['Day']=Time[1]
df['Day']=df['Day'].str.replace(",","")
df['Year']=Time[2]
```

```python
df.head()
```

```python
#Droping last updated columns
df.drop('Last Updated',axis=1,inplace=True)
df.head()
```

```python

df['Month'].unique()
```

```python
#Label encoding of month
df['Month']=label_encoder.fit_transform(df['Month'])
```

```python
df.head()
```

```python
df.dtypes
```

```python
#converting Day and year to int
df['Day']=df['Day'].apply(lambda x: int(x))
df['Year']=df['Year'].apply(lambda x: int(x))
```

```python
df.dtypes
```

```python
df.head()
```

```python
df.drop(['Current Ver','Android Ver','Category','Genres'],axis=1, inplace=True)
```
